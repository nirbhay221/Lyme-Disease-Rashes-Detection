```
import tensorflow as tf
import matplotlib.pyplot as plt 
import numpy as np
import zipfile as zipfile
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.image as mpimg
import csv 
import os
```

```
local_dir = '/content/archive (37).zip'
zip_ref = zipfile.ZipFile(local_dir,'r')
zip_ref.extractall('/tmp/')
zip_ref.close()
```

```
train_dir = '/tmp/RashData/Train/Train_2_Cases/'
train_lyme_pos= os.path.join(train_dir,'Lyme_Positive')
train_lyme_neg = os.path.join(train_dir,'Lyme_Negative')
```

```
lyme_pos_files = os.listdir(train_lyme_pos)
lyme_neg_files = os.listdir(train_lyme_neg)
```

```
pic_index=2
next_lyme_pos = [os.path.join(train_lyme_pos,fname) for fname in lyme_pos_files[pic_index-2:pic_index]]
next_lyme_neg = [os.path.join(train_lyme_neg,fname) for fname in lyme_neg_files[pic_index-2:pic_index]]
```

```
for i,img_path in enumerate(next_lyme_pos+next_lyme_neg):
  img= mpimg.imread(img_path)
  plt.imshow(img)
  plt.axis('Off')
  plt.show()
```

```
train_datagen = ImageDataGenerator(rescale=1.0/255.,shear_range=0.2,width_shift_range=0.2,zoom_range=0.2,height_shift_range=0.2,fill_mode='nearest',horizontal_flip=True)
```

```
test_datagen = ImageDataGenerator(rescale=1.0/255.)
```

```
test_dir = '/tmp/RashData/Validation/Validation_2_Cases'

test_lyme_pos= os.path.join(test_dir,'Lyme_Positive')
test_lyme_neg = os.path.join(test_dir,'Lyme_Negative')
```

```
train_data_generator = train_datagen.flow_from_directory(train_dir,target_size=(220,220),class_mode='binary',batch_size=32)
test_data_generator = test_datagen.flow_from_directory(test_dir,target_size=(220,220),class_mode='binary',batch_size=32)
```

```
model = tf.keras.models.Sequential([
                                    tf.keras.layers.Conv2D(64,(3,3),activation ='relu',input_shape=(220,220,3)),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(64,(3,3),activation ='relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(128,(3,3),activation ='relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Conv2D(256,(3,3),activation ='relu'),
                                    tf.keras.layers.MaxPooling2D(2,2),
                                    tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dropout(0.9),
                                    tf.keras.layers.Dense(400,activation='relu'),
                                    tf.keras.layers.Dense(1,activation='sigmoid')

])
```

```
model.compile(loss= 'binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```
history=model.fit(train_data_generator,epochs=50,validation_data=test_data_generator)
```

```

acc= history.history['accuracy']
loss= history.history['loss']
epochs= range(len(acc))
val_acc=history.history['val_accuracy']
val_loss= history.history['val_loss']
plt.figure(figsize=(10,8))
plt.plot(epochs,acc,'r')
plt.plot(epochs,val_acc,'b')
plt.title('accuracy and val_accuracy')
plt.show()
plt.figure(figsize=(10,8))
plt.plot(epochs,loss,'r')
plt.plot(epochs,val_loss,'b')
plt.title('loss and val_loss')
plt.show()
```
